# Persentation / Demo for my module : Co-ordinate geometry

```python
from d2_geometry import Cartesian_2D as d2 
from d2_geometry import angle_between
from shapes import Triangle, Quadilateral

# both are same just difference is that one is a class and another is a function
from draw import Draw, draw_polygon

import matplotlib.pyplot as plt

plt.style.use('bmh')
```

```python
# Quadilateral -> Rhombus
a = d2(2,-1)
b = d2(3,4)
c = d2(-2,3)
d = d2(-3,-2)

rhombus = Quadilateral(a,b,c,d)
print(rhombus.area(), 'sq.units')
print('This Quadilateral is', rhombus.which_quadilateral())
rhombus.draw(grid=False)
```

```python
# Triangle -> Right Angled, Scalene Triangle
a = d2(4,7)
b = d2(4,3)
c = d2(7,3)

t = Triangle(a,b,c)
print(t.area(), 'sq.units')
print(f'This is a {t.which_triangleByAngle()} Angled and {t.which_triangleBySide()} Triangle.')
t.draw(marker='v', grid=False)
```

```python
# draw
draw = Draw()

a = d2(3,3)
b = d2(6,4)
c = d2(8,7)
d = d2(5,6)

draw.polygon(a)
draw.polygon(a,b)
```
